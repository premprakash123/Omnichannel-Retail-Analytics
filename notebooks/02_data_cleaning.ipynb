```python
import pandas as pd
import numpy as np

# 1. Load the datasets
df_pos = pd.read_csv('../data/pos_sales.csv')
df_online = pd.read_csv('../data/online_sales.csv')

# 2. Combine them into one Master Dataframe
# This is the "Omnichannel" step!
df = pd.concat([df_pos, df_online], ignore_index=True)

print("Datasets merged. Total rows:", len(df))
df.head()
```

```python
# Check for NULL values in every column
print(df.isnull().sum())
```

```python
# We can see that 'order_id' has missing values.
# We cannot have a sale without an ID, so we drop those specific rows.
df = df.dropna(subset=['order_id'])

print("NULLs handled. New row count:", len(df))
df.isnull().sum()
```

```python
# Check how many duplicate rows exist
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")
```

```python
# Remove the duplicates
df = df.drop_duplicates()

print("Duplicates removed. Final row count:", len(df))
```

```python
df.shape
```

```python
# Convert timestamp column to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])
print(df['timestamp'].dtype) 
```

```python
df['price'].describe()
```

```python
outliers = df[df['price'] > 10000]
print("Outliers found:", len(outliers))
```

```python


df = df[df['price'] <= 10000]

df['price'].describe()
```

```python

print("Outliers removed. Data is now clean.")
```

```python
df.shape
```

```python
# Calculate total sales for each row
df['sales'] = df['quantity'] * df['price']

# Preview the result
print(df[['quantity', 'price', 'sales']].head())
```

```python
# Extract Month Name for seasonality analysis
df['month'] = df['timestamp'].dt.month_name()

# Extract Year for YoY analysis
df['year'] = df['timestamp'].dt.year

# Preview the result
print(df[['timestamp', 'year', 'month']].head())
```

```python
# Save the cleaned master dataset
df.to_csv('../data/omnichannel_cleaned_master.csv', index=False)
print("Cleaned master dataset saved as omnichannel_cleaned_master.csv")
```
