```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Expanded product list for more realistic analysis
products = {
    'Electronics': {'price': 500, 'weight': 1.0},
    'Clothing': {'price': 40, 'weight': 0.6},
    'Home Decor': {'price': 120, 'weight': 0.8},
    'Beauty': {'price': 60, 'weight': 0.4},
    'Grocery': {'price': 20, 'weight': 1.2}
}
cities = ['Bengaluru', 'Mumbai', 'Delhi', 'Hyderabad', 'Chennai', 'Pune', 'Kolkata']

def generate_professional_data(filename, rows=10000, source="POS"):
    data = []
    start_date = datetime(2023, 1, 1)
    
    for i in range(rows):
        # 1. Intentionally adding missing values (5% chance)
        order_id = np.nan if random.random() < 0.05 else f"{source}_{100000 + i}"
        
        product_cat = random.choice(list(products.keys()))
        base_price = products[product_cat]['price']
        price = base_price * random.uniform(0.8, 1.2)
        qty = random.randint(1, 5)
        
        # 2. Generate Date with SEASONAL BIAS (Q4 Spike)
        # We pick a random day in the last 450 days
        days_offset = random.randint(0, 450)
        date = start_date + timedelta(days=days_offset, hours=random.randint(0, 23))
        
        # Logic: If it's November (11) or December (12), increase the chance of this row existing
        # This ensures that when you plot a line chart, Q4 looks higher.
        if date.month in [11, 12] and random.random() < 0.3:
            # We "duplicate" some Q4 entries to simulate a rush
            data.append([order_id, product_cat, qty, price, date, random.choice(cities)])
        
        city = random.choice(cities)
        data.append([order_id, product_cat, qty, price, date, city])
    
    df = pd.DataFrame(data, columns=['order_id', 'product', 'quantity', 'price', 'timestamp', 'city'])
    
    # 3. Add duplicate records (about 2% of data)
    duplicates = df.sample(frac=0.02)
    df = pd.concat([df, duplicates], ignore_index=True)
    
    # 4. Add some "extreme outliers" in price (required for the cleaning phase)
    # Change 10 random prices to something crazy like 99,999
    random_indices = df.sample(n=10).index
    df.loc[random_indices, 'price'] = 99999.0
    
    df.to_csv(filename, index=False)
    print(f"Successfully generated {len(df)} rows for {filename}")

# Generate 10,000 rows for each source
generate_professional_data('pos_sales.csv', rows=10000, source="POS")
generate_professional_data('online_sales.csv', rows=10000, source="ONLINE")
```
